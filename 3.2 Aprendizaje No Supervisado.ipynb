# <center> Aprendizaje Supervisado

El aprendizaje no supervisado es una rama del aprendizaje automático en la que se utilizan algoritmos para descubrir patrones, estructuras o características ocultas en los datos sin la necesidad de etiquetas o respuestas previas. A diferencia del aprendizaje supervisado, en el que se tiene un conjunto de datos etiquetados, en el aprendizaje no supervisado los datos no están etiquetados y el objetivo principal es explorar y comprender la estructura subyacente de los datos.

Los algoritmos de aprendizaje no supervisado se utilizan en una amplia gama de aplicaciones, como la segmentación de clientes, la detección de anomalías, la reducción de dimensionalidad, la agrupación de datos, la generación de recomendaciones, el análisis de texto, entre otros.

A continuación, se presentan algunas de las principales bibliotecas en Python que se utilizan para desarrollar proyectos de aprendizaje no supervisado:

1. Scikit-learn: Es una biblioteca popular de aprendizaje automático que proporciona una amplia gama de algoritmos tanto para aprendizaje supervisado como no supervisado. Incluye implementaciones de algoritmos de agrupación (clustering) como K-means, DBSCAN y clustering espectral, así como algoritmos de reducción de dimensionalidad como PCA y t-SNE.

2. TensorFlow: Es una biblioteca de aprendizaje automático de código abierto desarrollada por Google. Aunque es conocida principalmente por su enfoque en el aprendizaje profundo (deep learning), también incluye algoritmos de aprendizaje no supervisado, como Autoencoders y algoritmos de clustering como el algoritmo K-means.

3. PyTorch: Es otra biblioteca de aprendizaje automático popular, especialmente utilizada para el desarrollo de modelos de aprendizaje profundo. También proporciona algoritmos de aprendizaje no supervisado, como Autoencoders y algoritmos de clustering como K-means.

4. NumPy: Es una biblioteca fundamental en Python para el cálculo numérico. Proporciona una serie de funciones y estructuras de datos eficientes que son utilizadas en muchos algoritmos de aprendizaje no supervisado.

4. pandas: Es una biblioteca de manipulación y análisis de datos en Python. Proporciona estructuras de datos flexibles para trabajar con datos tabulares y tiene funciones que facilitan la preparación y limpieza de datos antes de aplicar algoritmos de aprendizaje no supervisado.

5. NLTK (Natural Language Toolkit): Es una biblioteca de procesamiento de lenguaje natural en Python. Proporciona herramientas y recursos para trabajar con datos de texto y aplicar técnicas de aprendizaje no supervisado en problemas de análisis de texto, como clustering de documentos y detección de temas.

En cuanto a los ejemplos de la vida real, aquí tienes algunos casos de uso comunes del aprendizaje no supervisado:

1. Segmentación de clientes: Utilizando algoritmos de clustering, se pueden agrupar los clientes según sus características y comportamiento para identificar segmentos específicos y personalizar estrategias de marketing.

2. Detección de anomalías: Se pueden aplicar algoritmos de detección de anomalías para identificar puntos de datos atípicos en conjuntos de datos, lo que puede ser útil para detectar fraudes

## Ejemplo útil

Crea un ejemplo en python donde los datos sean un data frame de 100 personas y segmentes por cantidad de productos comprados y cantidad para darles un incentivo a dichos clientes con algun algoritmo de aprendizaje no supervisado

Para este ejemplo vamos a utilizar el algoritmo de clustering K-Means para segmentar a los clientes según la cantidad de productos comprados y la cantidad de productos comprados para ofrecerles un incentivo. Para esto, vamos a generar un dataset ficticio con 100 personas y sus compras.

Primero, importamos las librerías necesarias y generamos el dataset aleatorio:

```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

np.random.seed(0)
n_samples = 100

# Generamos los datos aleatorios
productos_comprados = np.random.randint(0, 50, n_samples)
cantidad_comprada = np.random.randint(0, 5, n_samples)
datos = pd.DataFrame({'Productos Comprados': productos_comprados, 'Cantidad Comprada': cantidad_comprada})

```

Luego, escalamos los datos para que tengan media cero y varianza unitaria:

```python
# Escalamos los datos
scaler = StandardScaler()
datos_escalados = scaler.fit_transform(datos)
```

A continuación, aplicamos el algoritmo de clustering K-Means con dos clusters y obtenemos las etiquetas de cada punto:

```python
# Aplicamos el algoritmo de clustering K-Means
kmeans = KMeans(n_clusters=2, random_state=0).fit(datos_escalados)
etiquetas = kmeans.labels_
```

Finalmente, podemos agregar las etiquetas al dataset y ver cómo se segmentan los clientes según la cantidad de productos comprados y la cantidad para darles un incentivo:

```python
# Agregamos las etiquetas al dataset
datos['Cluster'] = etiquetas

# Mostramos el resultado de la segmentación
print(datos.groupby(['Cluster', 'Cantidad Comprada']).agg({'Productos Comprados': 'mean'}))
```

Este código agrupa los datos según el cluster al que pertenecen y la cantidad de productos comprados. Luego, muestra la media de productos comprados para cada grupo. Podemos usar esta información para ofrecer incentivos específicos a cada grupo de clientes.

Este es un ejemplo sencillo de cómo se puede utilizar el aprendizaje no supervisado en un contexto de negocios para segmentar a los clientes y ofrecer incentivos personalizados.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Generar datos aleatorios para el ejemplo
np.random.seed(0)
num_muestras = 1000
num_caracteristicas = 2

# Generar datos aleatorios en forma de dos círculos
X, y = make_circles(n_samples=num_muestras, noise=0.05, factor=0.5, random_state=0)

# Crear y entrenar el modelo de K-means
modelo_kmeans = KMeans(n_clusters=2)
modelo_kmeans.fit(X)

# Realizar predicciones en los datos
predicciones = modelo_kmeans.predict(X)

# Calcular el coeficiente de silueta del modelo
coeficiente_silueta = silhouette_score(X, predicciones)

# Graficar los resultados
plt.scatter(X[:, 0], X[:, 1], c=predicciones, cmap='viridis')
plt.scatter(modelo_kmeans.cluster_centers_[:, 0], modelo_kmeans.cluster_centers_[:, 1], marker='x', color='red', s=100, label='Centroides')
plt.title('Clustering K-means')
plt.xlabel('Característica 1')
plt.ylabel('Característica 2')
plt.legend()
plt.show()

print("Coeficiente de silueta:", coeficiente_silueta)
```

En este ejemplo, utilizamos el algoritmo de clustering K-means mediante KMeans() de Scikit-learn. Generamos datos aleatorios utilizando la función make_circles() para crear dos círculos superpuestos.

Luego, creamos y entrenamos el modelo de K-means con n_clusters=2, lo que significa que esperamos dos clusters en los datos. Realizamos predicciones en los datos utilizando predict().

Después, calculamos el coeficiente de silueta (silhouette_score()) del modelo para evaluar la calidad de los clusters obtenidos. Un coeficiente de silueta cercano a 1 indica una buena separación entre los clusters.

Finalmente, graficamos los datos y los clusters obtenidos, utilizando plt.scatter(). También graficamos los centroides de los clusters (modelo_kmeans.cluster_centers_) en rojo.

Recuerda que este ejemplo utiliza datos generados aleatoriamente y es solo un caso de muestra para demostrar el uso de aprendizaje no supervisado. En un escenario real, deberías utilizar datos reales y ajustar los hiperparámetros del modelo para obtener resultados más precisos y significativos.

## Diferencias

La principal diferencia entre el aprendizaje supervisado y el aprendizaje no supervisado radica en la disponibilidad de etiquetas o respuestas conocidas en los datos de entrenamiento.

En el aprendizaje supervisado, se tienen datos de entrenamiento que están etiquetados, lo que significa que cada muestra tiene una etiqueta o respuesta conocida. El objetivo del modelo es aprender una función o relación entre las características de entrada y las etiquetas para poder hacer predicciones precisas en datos nuevos. El modelo se entrena utilizando ejemplos de entrenamiento, donde las características y las etiquetas están asociadas. Algunos ejemplos de algoritmos de aprendizaje supervisado son la regresión lineal, las máquinas de vectores de soporte (SVM) y las redes neuronales.

En el aprendizaje no supervisado, por otro lado, no se tienen etiquetas o respuestas conocidas en los datos de entrenamiento. El objetivo principal es explorar y encontrar patrones o estructuras ocultas en los datos sin la guía de las etiquetas. El modelo se entrena únicamente en función de la estructura inherente de los datos y se utiliza para descubrir relaciones, agrupaciones o características latentes. Algunos ejemplos de algoritmos de aprendizaje no supervisado son el clustering (como K-means y DBSCAN), la reducción de dimensionalidad (como PCA y t-SNE) y la detección de anomalías.

En resumen, la diferencia clave entre el aprendizaje supervisado y el no supervisado radica en si los datos de entrenamiento tienen etiquetas conocidas o no. En el aprendizaje supervisado, el modelo aprende a partir de ejemplos etiquetados para hacer predicciones precisas, mientras que en el aprendizaje no supervisado, el modelo descubre patrones o estructuras ocultas en los datos sin la necesidad de etiquetas previas.
